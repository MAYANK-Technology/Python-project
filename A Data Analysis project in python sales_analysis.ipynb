```python
import pandas as pd
import plotly.express as px

# Sample DataFrame
# Assuming 'sales' is your DataFrame
# Example: sales = pd.DataFrame({...})

# Select only numeric columns
numeric_sales = sales.select_dtypes(include=[float, int])

# Calculate the correlation matrix
corr = numeric_sales.corr()

# Create an interactive heatmap
fig = px.imshow(corr,
                text_auto=True,
                aspect="auto",
                color_continuous_scale='RdBu')

# Update the layout to position the labels at the top and left
fig.update_layout(
    xaxis=dict(side="top"),  # Move x-axis labels to the top
    yaxis=dict(side="left")  # Move y-axis labels to the left
)

fig.show()


```
